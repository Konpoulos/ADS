```python
#convert id to int64
movies_metadata = movies_metadata.drop([19730, 29503, 35587])
movies_metadata['id'] = movies_metadata['id'].astype("int")

#Combine the datasets

merged_movies = movies_metadata.merge(movies_ratings, left_on='id', right_on='movieId')
merged_movies
```

```python
merged_movies[merged_movies['id'] == 2].head(100)
```

```python
from itertools import permutations

# create a function that makes combinations of the movies that a user (x) reviewed
def create_combinations(x):
    combinations = pd.DataFrame(list(permutations(x.values, 2)), columns=['Movie_a', 'Movie_b'])
    return combinations
```

```python
avg_user = merged_movies.groupby(['userId']).size().reset_index(name='counts')
avg_user = avg_user.set_index('userId')
avg_user.reset_index().sort_values('counts', ascending=False)
```

```python
avg_user = avg_user[avg_user['counts'] > 30]
users_index = avg_user.index

smaller_selection = merged_movies.iloc[users_index]
smaller_selection
```

```python
smaller_selection.shape
```

```python
# use the create_combinations function

Movies_combinations = smaller_selection.groupby('userId')['original_title'].apply(create_combinations)

Movies_combinations.head(10)
```

```python
# calculate the occurences of book a and book b

Movies_combinations = Movies_combinations.reset_index(drop=True)
combi_count = Movies_combinations.groupby(['Movie_a', 'Movie_b']).size()
combi_count = combi_count.to_frame(name='size').reset_index()
combi_count.sort_values('size', ascending=False).head(5)
```
